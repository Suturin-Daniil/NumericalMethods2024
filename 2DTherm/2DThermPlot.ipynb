```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation 
```

```python
GridInformation = np.loadtxt('2DThermSolution.txt', max_rows=1, delimiter=' ', dtype=object)
Nx = int(GridInformation[1])
Ny = int(GridInformation[3])
hx = float(GridInformation[5])
hy = float(GridInformation[7])
Data = np.loadtxt('2DThermSolution.txt', skiprows=1, delimiter=' ')
Data[0].reshape(Nx,Ny), Data[0]
```

```python
x = np.linspace(hx,1-hx,Nx)
y = np.linspace(hy,1-hy,Ny)

X, Y = np.meshgrid(x,y)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
#cm = plt.get_cmap('inferno')
a = ax.contourf(X,Y,Data[10].reshape(Ny, Nx), levels=400, cmap='jet', vmin=10, vmax=100)
cbar = fig.colorbar(a,)
#cbar.set_clim(10,100)

fig1 = plt.figure(figsize=(8,8))
ax1 = fig1.add_subplot(projection = '3d')
ax1.plot_surface(X,Y,Data[30].reshape(Ny, Nx), cmap = 'inferno', vmin=10, vmax=120)
```

```python
def anima(i):
    ax1.clear()
    #ax.clear()
    #fig.clear()
    #ax.contourf(X,Y,Data[i].reshape(Ny, Nx), levels=400, vmin=10, vmax=100, extend='both')
    ax1.plot_surface(X,Y,Data[i].reshape(Ny, Nx), cmap = 'jet', vmin=10, vmax=101)
    ax1.set_zlim((10,101))
    return fig1,
#fig, ax = plt.subplots(figsize=(8,8))
ani = FuncAnimation(fig1, anima, frames=100, interval=50)
ani.save('2DTherm.gif',writer='pillow',fps=100,dpi=80)
```

```python
infZ = np.loadtxt('2DThermSolutionZadavalnik.txt', max_rows=1, dtype='object', delimiter=' ')
Nx = int(infZ[1])
Ny = int(infZ[3])
hx = float(infZ[5])
hy = float(infZ[7])
datZ = np.loadtxt('2DThermSolutionZadavalnik.txt', skiprows=1, delimiter=' ')

x = np.linspace(hx,1-hx,Nx)
y = np.linspace(hy,1-hy,Ny)

X, Y = np.meshgrid(x,y)
```

```python
fig1 = plt.figure(figsize=(12,12))
ax1 = fig1.add_subplot(projection = '3d')
ax1.clear()
def a(i):
    ax1.clear()
    ax1.plot_surface(X, Y, datZ[i].reshape(Ny,Nx), cmap = 'jet', vmin=0, vmax=1)
    ax1.set_zlim((0,1.25))
    return fig1
an = FuncAnimation(fig1, a, frames = len(datZ)-1, interval=15)
an.save('2DThermZadavalnik.gif',writer='pillow',fps=25,dpi=100)
```
